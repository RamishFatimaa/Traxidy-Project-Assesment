```python
import pandas as pd
import numpy as np
df=pd.read_csv("gyr_constraints.csv")
df
```

```python
unique_statuses = df['status'].unique()
print(unique_statuses)
```

```python
status_mapping = {'green': 1, 'yellow': 2, 'red': 3}

# Create a new column with numeric values
df['num_status'] = df['status'].map(status_mapping)

# Display the updated DataFrame
print(df)
```

```python
status_counts = df['num_status'].value_counts()
print(status_counts)
```

```python
# Save the DataFrame to a CSV file
df.to_csv('update_gyr_constraints.csv', index=False)

print("CSV file saved as 'updated_status.csv'.")
```
